```python
import pandas as pd

evi = pd.read_csv('../data/satdata/evi.csv', parse_dates=[0], index_col=[0]).interpolate()
stemp = pd.read_csv('../data/satdata/stemp.csv', parse_dates=[0], index_col=[0]).interpolate()
df = pd.concat([evi, stemp], axis=1, keys=["evi", "stemp"])
display(df.head(2))
df_year = df.groupby(df.index.year).mean()
df_year = df_year.interpolate()
display(df_year.head(2))
df_res = df_year.unstack().reset_index()
df_res.columns = ['type', 'buurt', 'year', 'value']
df_res = df_res.pivot_table(index=['year','buurt'], columns=['type'], values='value').reset_index()
df_res
```

# Temperatuur
De temperatuur in de buurt is van groot belang op het leefcomfort, daarnaast kan het in warme perioden leiden tot negatieve effecten op de gezondheid en welzijn van de inwoners. Groen kan bijdragen aan het beperken van de opwarming van de stad en aan het vergroten van het thermisch comfort; hieronder laten we zien dat meer groen leidt tot een lagere oppervlakte temperatuur.

Voor de oppervlakte temperatuur (surface temperature) is een satelliet product gebruikt van [MODIS](https://developers.google.com/earth-engine/datasets/catalog/MODIS_006_MOD11A1), de gegevens van 2000 tot 2020 zijn verzameld met [Google Earth Engine](https://developers.google.com/earth-engine).

Hieronder plotten we de gemiddelde temperatuur waarden voor de 54 wijken voor de 20 jaren (2000-2020), ten opzichte van de gemiddelde ["EVI"](https://en.wikipedia.org/wiki/Enhanced_vegetation_index) (de groen index). Hiermee wordt per jaar de relatie tussen groen en de temperatuur duidelijk. 

\* Door per jaar te kijken worden de klimatologische verschillen (tussen verschillende jaren) in de groenindex en de temperatuur zoveel mogelijk uitgesloten.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

plt.figure(figsize=(11, 9))
sns.lmplot("evi", "stemp", data=df_res, hue="year", truncate=True)
plt.title("Temperatuur en EVI-index voor 20 jaar, voor 54 wijken in Breda", fontsize=14)
plt.xlabel("EVI", fontsize=12)
plt.ylabel("Temperatuur in graden Celsius", fontsize=12)
plt.show()
```

## Dieper in de data

Zoals hieronder is te zien schommelt de gemiddelde temperatuur van een jaar in Breda zo rond de 15-17 graden Celsius de afgelopen 20 jaar, er is geen duidelijk trend in terug te zien. De temperatuur per buurt is daaronder weergegeven er zijn duidelijke verschillen tussen de buurten te zien, erg groene buurten zoals bossen en buitengebieden (oa. Mastbos, Buitengebied Ulvenhout) hebben lagere temperaturen dan stedelijke, grijzere gebieden zoals oa. Station, City en Belcrum, het verschil tussen deze buurten is wel tot 4 graden Celsius! De temperatuur in groenere wijken blijkt ook een stuk constanter te zijn dan in grijzere wijken.

```python
stemp_average_year = df_res.groupby('year').mean()['stemp']

sns.set()
plt.figure(figsize=(11, 5))
sns.barplot(stemp_average_year.index, stemp_average_year.values)
plt.title("Gemiddelde temperatuur per jaar in Breda", fontsize=20)
plt.xlabel("Jaar", fontsize=15)
plt.ylabel("Gem. temperatuur Breda in Celsius", fontsize=15)
plt.show()
```

```python
df_temp_buurt = df['stemp'].unstack().reset_index()
df_temp_buurt.columns = ['buurt', 'datum', 'temp']
stemp_average_buurt = df_res.groupby('buurt').mean()['stemp'].sort_values()
sorter = stemp_average_buurt.index
df_temp_buurt.buurt = df_temp_buurt.buurt.astype('category')
df_temp_buurt.buurt.cat.set_categories(sorter, inplace=True)
df_temp_buurt.sort_values(['buurt']) # sorted based on average temp over 20 years

plt.figure(figsize=(9, 12))
sns.boxplot(x='temp', y='buurt', data=df_temp_buurt, orient="h")
plt.title("Boxplot temperatuur per buurt (54) in Breda", fontsize=20)
plt.xlabel("Temperatuur in Celsius", fontsize=15)
plt.ylabel("Buurt", fontsize=15)
plt.show()
```

## EVI
Het klimatologisch verschil tussen jaren wordt meegenomen door jaren als variabele mee te nemen in het regressie model, hiermee blijkt evi een significant aandeel heeft; EVI heeft een coeficient van -0.0012 (p<0.000). Dat betekend dat een wijk die met 1000 evi punten vergroent gemiddeld 1.2 graden koeler wordt.

```python
from statsmodels.formula.api import ols

fit = ols("stemp ~ evi + C(year) ", data=df_res).fit()
fit.summary()
```
